### Step 7. Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)

```python
def majorty(x):
  if x>17:
    return True
  else:
    return False
```

```python
def time10 (x):
  if type(x) in [int, float]:
    return 10*x
  else:
    return x
```
