What Is the Average Height of US Presidents?
Use file president_heights.csv, which is a comma-separated list of labels and values

### Step1:
Read the data with pandas, turn the 'height(cm)' column into a numpy array. Print the array.

```python
import pandas as pd
import numpy as np
mydata=pd.read_csv('president_heights.csv')
```

```python
mydata.columns
```

```python
mydata.head
```

```python
heights = np.array(mydata['height(cm)'])
```

### Step2:
- Print the mean of the heights
- Print the standard deviatoin of heights
- Print the minimum height
- Print the maximum height

```python
heights.mean()
```

```python
heights.std()
```

```python
heights.min()
```

```python
heights.max()
```

```python
mydata['height(cm)'].max()
```

```python
mydata.columns
```

### Step3:
- Print the 25 percentile
- Print the median of heights
- Print the 75 percentile


```python
np.percentile(heights, 25)
```

```python
np.percentile(heights, 50)
```

```python
np.percentile(heights, 75)
```

### Step4:
- Create a histogram of the heights
- Give a title to the histogram
- Give x and y labels


```python
import matplotlib.pyplot as plt
plt.hist(heights, bins=10)
plt.title('heights distribution of US presidents')
plt.xlabel('heights(cm)')
plt.ylabel('numbers')
```
